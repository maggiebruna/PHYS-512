```python
#solving for the decay products of U238

import numpy as np
import matplotlib.pyplot as plt
from scipy import integrate
import time
```

```python
#defining the half life and decay rate 
hl = np.array([4.468e9, 0.066, 7.65e-4, 245500, 75380, 1600, 0.0105, 5.90e-6, 5.09e-5, 3.79e-5, 5.21e-12, 22.3, 5.015, 0.379])
dec_rate = np.log(2)/hl


#populating a 15x15 matrix for the decay rate 
dec_matrix = np.zeros((len(hl)+1,len(hl)+1))    
np.fill_diagonal(dec_matrix,-dec_rate)
np.fill_diagonal(dec_matrix[1:],dec_rate[:-1])
#defining our y matrix with an initial uranium value

y_start = np.zeros(len(hl)+1)
y_start[0] = 1 

def func(t,y):
    dydx = dec_matrix@y
    return dydx

x = [0,16e9]
solution = integrate.solve_ivp(func,x,y_start,method='Radau')
y = solution.y
time = solution.t
```

```python
#part b, plotting a bunch of stuff! 

#plotting a ratio of Pb206 to U238 

U238 = y[0]
Pb206 = y[14]
Th234 = y[1]
U234 = y[3]
```
